```python
def analyse_video(video_path):
  cap = cv2.VideoCapture(video_path)
  # identify objects in each frame
  ret = True
  frames = []
  # read frames
  while ret:
      ret, frame = cap.read()
      if ret:
          data = read_frame(frame)
          frames.append(data)
          if cv2.waitKey(25) & 0xFF == ord('q'):
            #print("done")
            break
```
